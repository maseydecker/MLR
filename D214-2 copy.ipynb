```python
#cleaning up the levels column
levels_group= [(df['levels'].str.contains('One Story', case=False, regex=False), 'One'), 
               (df['levels'].str.contains('Two Story', case=False, regex=False), 'Two'),
               (df['levels'].str.contains('Three or More', case=False, regex=False), 'Three or More'),
               (df['levels'].str.contains('One-Two', case=False, regex=False), 'Two'),
               (df['levels'].str.contains('Multi/Split', case=False, regex=False), 'Multi/Split'),
               (df['levels'].str.contains('0', case=False, regex=False), 'Zero'),
               (df['levels'].str.contains('Three', case=False, regex=False), 'Three or More'),
               (df['levels'].str.contains('2', case=False, regex=False), 'Two'),
               (df['levels'].str.contains('Tri-Level', case=False, regex=False), 'Tri-Level'),
               (df['levels'].str.contains('1', case=False, regex=False), 'One'),
               (df['levels'].str.contains('3', case=False, regex=False), 'Three or More'),
               (df['levels'].str.contains('Split Level', case=False, regex=False), 'Multi/Split'),
               (df['levels'].str.contains('Other-One', case=False, regex=False), 'One'),
               (df['levels'].str.contains('4+', case=False, regex=False), 'Three or More'),
               (df['levels'].str.contains('Multi-Level', case=False, regex=False), 'Multi/Split'),
               (df['levels'].str.contains('Two-One', case=False, regex=False), 'Two')]
```

```python
levels_criteria, levels_values = zip(*levels_group)
df['levels_Group'] = np.select(levels_criteria, levels_values, 'other')
```

```python
#split data in test and train
y = df['price']
X = df.drop('price', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size = 0.3, random_state = 1)
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.summary()
```

```python
import seaborn as sns
y_train_price = lr_1.predict(X_train_lm)
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

```python
X_test_m = sm.add_constant(X_test)

# Making predictions using the final model
y_pred_m = lr_1.predict(X_test_m)
```

```python
from sklearn.metrics import r2_score
r2_score(y_true = y_test, y_pred = y_pred_m)
```

```python
from sklearn import metrics

meanAbErr = metrics.mean_absolute_error(y_test, y_pred_m)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_m)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_m))

print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
import pylab

# check for the normality of the residuals
sm.qqplot(lr_1.resid, line='s')
pylab.show()

# also check that the mean of the residuals is approx. 0.
mean_residuals = sum(lr_1.resid)/ len(lr_1.resid)
print("The mean of the residuals is {:.4}".format(mean_residuals))
```
